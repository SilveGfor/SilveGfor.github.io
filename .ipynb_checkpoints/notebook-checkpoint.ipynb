```python
genders = list(df['gender'].unique())
genders_values = len(df[df['gender'] == genders[0]]), len(df[df['gender'] == genders[1]])
pie_fig = go.Figure(data=[go.Pie(labels=genders, values=genders_values)])
pie_fig.show()
pie_fig.write_html("gr1_1.html")
```

```python
senior_citizen = list(df['SeniorCitizen'].unique())
senior_citizen_values = len(df[df['SeniorCitizen'] == senior_citizen[0]]), len(df[df['SeniorCitizen'] == senior_citizen[1]])
pie_fig = go.Figure(data=[go.Pie(labels=senior_citizen, values=senior_citizen_values, pull=[0, 0.15], hole=0.9)])
pie_fig.update_layout(
    annotations=[dict(text='The attitude of<br>young clients towards<br>the elderly', x=0.5, y=0.5, font_size=20, showarrow=False)])

pie_fig.show()
pie_fig.write_html("gr1_2.html")
```

```python
fig_hist = go.Figure(data=[go.Histogram(x=df.tenure)])
fig_hist.update_layout(
    xaxis_title="Number of months the customer has stayed with the company",
    yaxis_title="Number of clients")
fig_hist.show()
fig_hist.write_html("gr1_3.html")
```

```python
labels = list(df.Contract.value_counts().index)
values = list(df.Contract.value_counts().values)

fig = go.Figure(data=[go.Bar(x = labels, y = values)])
fig.update_layout(
    title="The contract term of the customer",
    title_x = 0.5,
    xaxis_title="Number of months the customer has stayed with the company",
    yaxis_title="Number of clients")
fig.show()
fig.write_html("gr1_4.html")
```

```python
out_df = df[df['Churn'] == 'Yes']['TotalCharges']
remain_df = df[df['Churn'] == 'No']['TotalCharges']

fig = go.Figure()
fig.add_trace(go.Box(y=out_df, name='churned customers'))
fig.add_trace(go.Box(y=remain_df, name='remaining customers'))
fig.update_layout(title="Comparison of the amounts of all payments of churned and remaining customers")
fig.show()
fig.write_html("gr1_5.html")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer

from xgboost import XGBClassifier
```

```python
X = df.drop(["customerID", "Churn"], axis=1)
y = df[["Churn"]]

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.20, random_state = 4242)
```

**Градиентный бустинг из sklearn**

```python
### Загружаем библиотеки и данные
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import mean_squared_error,r2_score

import warnings
warnings.filterwarnings('ignore')
breast_cancer = load_breast_cancer()

params = {
          # 'n_estimators':100,
          # 'max_depth':6,
          # 'criterion':'friedman_mse',
          # 'learning_rate':0.3,
          # 'min_samples_leaf':16,
          # 'min_samples_split':16
          }

gbr = GradientBoostingRegressor(**params)
gbr.fit(X_train,y_train)

clf = GradientBoostingClassifier(**params).fit(X_train, y_train)

train_accuracy_score=clf.score(X_train,y_train)
print(train_accuracy_score)

test_accuracy_score=clf.score(X_test,y_test)
print(test_accuracy_score)
```

**Градиентный бустинг из XGBoost**

```python
pipeline = make_pipeline(
                         XGBClassifier(
                            #  n_estimators=100, # This is the total number of trees that are trained
                            #  learning_rate=0.3, # The learning rate used to combine the result of subequent trees
                            #  max_depth=6, # The maximum depth of each tree
                            #  subsample=1.0, # The proportion of the dataset used to build each tree (e.g., 0.5 uses 50% of the dataset selected randomly)
                            #  colsample_bynode=1.0, # The proportion of features allowed to be used to make each split (e.g., 0.75 only considers 75% of the input features at random to make a split)
                            #  random_state=0 # This lets us get the same results on repeated executions
                         )
                        ).fit(X_train, y_train)
```

```python
(pipeline.predict(X_test).reshape(1409 ,1)==np.array(y_test)).sum()/float(len(X_test)) #acc
```
